```python
scenic_score = np.zeros(data.shape)
```

```python
for i in range(1,data.shape[0]-1):
    for j in range(1, data.shape[1]-1):
        value = data[i,j]
        left = data[i,:j]
        right = data[i,j+1:]
        top = data[:i,j]
        bottom = data[i+1:,j]
        
        
        if (value<= left).sum()>0:
            left_val = len(left)-(np.where(left >= value)[0].max())
        else:
            left_val = len(left)
        
        if (value<= top).sum()>0:
            top_val = len(top)-(np.where(top >= value)[0].max())
        else:
            top_val = len(top)
        if (value<= right).sum()>0:
            right_val =np.where(right >= value)[0].min()+1
        else:
            right_val = len(right)
        if (value<= bottom).sum()>0:
            bot_val =np.where(bottom >= value)[0].min()+1
        else:
            bot_val = len(bottom)
#         print(i,j,"", right_val,left_val,top_val,bot_val)
        ss = right_val*left_val*top_val*bot_val
        scenic_score[i,j] = ss
```
