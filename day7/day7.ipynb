```python
class file:
    def __init__(self,name, size):
        self.name = name
        self.size = int(size)
    def __str__(self):
        return f'{self.name} {self.size}'

class directory:
    def __init__(self, name, parent):
        self.name = name
        self.parent = parent
        self.contents = []
    def __str__(self):
        ret_str = self.name+" "+str(self.sum_contents())+'\n'
        for c in self.contents:
            ret_str+=str(c)+'\n'
        return ret_str
    def sum_contents(self):
        total = 0
        for c in self.contents:
            if type(c) == directory:
                total+= c.sum_contents()
            else:
                total+=c.size
        return total
    def get_small_folders(self):
        ret_list = []
        for c in self.contents:
            if type(c) == directory:
                ret_list+=c.get_small_folders()
                if c.sum_contents()<=100000:
                    ret_list.append(c.sum_contents())
    
        return ret_list
    def get_all_folders(self):
        ret = []
        for c in self.contents:
            if type(c) == directory:
                ret+=c.get_all_folders()
                ret.append([c.name, c.sum_contents()])
        return ret
```

```python
root_node = directory('root',None)
root_node.contents.append(directory('/', root_node))
```

```python
cur_node = root_node
data = data_orig.copy()
i=0
while i < len(data):
    print(cur_node)
    line = data[i]
    print(line)
    if line[0] == '$':
        command = line.split(" ")
        print(command)
        if command[1] == 'cd':
            if command[2] == '..':
                
                cur_node = cur_node.parent
                
            else:
                
                for item in cur_node.contents:
                    if type(item) == directory:
                        if item.name == command[2]:
                            cur_node = item
                
        if command[1] == 'ls':
            while data[i+1][0] != '$':
                print(data[i+1])
                new_line = data[i+1]
                if new_line[:3] == 'dir':
                    cur_node.contents.append(directory(new_line.split(" ")[1], cur_node))
                else:
                    cur_node.contents.append(file(new_line.split(" ")[1],new_line.split(" ")[0] ))
                i+=1
                
                #grow tree
            print(cur_node)
            pass
    i+=1
```
